```python
# ============================================================
# Import y clase base abstracta para simulacion(iconos)
#Definiciones
# ============================================================

from abc import ABC, abstractmethod

class Dispositivo(ABC):
    """
    Clase base que define la interfaz general para dispositivos inteligentes.
    """
    def __init__(self, id_dispositivo: str):
        self.__id_dispositivo = id_dispositivo
        self.__estado = False  # False = apagado, True = encendido

    @abstractmethod
    def encender(self):
        pass

    @abstractmethod
    def apagar(self):
        pass

    @abstractmethod
    def mostrar_datos(self):
        pass

    # MÃ©todos protegidos para acceder al estado e id
    def _set_estado(self, estado: bool):
        self.__estado = estado

    def _get_estado(self) -> bool:
        return self.__estado

    def _get_id(self) -> str:
        return self.__id_dispositivo
```

```python
# ============================================================
# LuzInteligente
# ============================================================

class LuzInteligente(Dispositivo):
    """
    Luz inteligente con control de intensidad (0-100%).
    """
    def __init__(self, id_dispositivo: str, intensidad: int = 0):
        super().__init__(id_dispositivo)
        self.intensidad = intensidad

    def encender(self):
        self._set_estado(True)
        self.intensidad = 100
        print(f"ğŸ’¡ Luz {self._get_id()} encendida al {self.intensidad}%.")

    def apagar(self):
        self._set_estado(False)
        self.intensidad = 0
        print(f"ğŸ’¡ Luz {self._get_id()} apagada.")

    def mostrar_datos(self):
        estado = "Encendida" if self._get_estado() else "Apagada"
        print(f"[LuzInteligente] ID: {self._get_id()} | Estado: {estado} | Intensidad: {self.intensidad}%")
```

```python
# ============================================================
# CamaraSeguridad
# ============================================================

class CamaraSeguridad(Dispositivo):
    """
    CÃ¡mara de seguridad con estado de grabaciÃ³n.
    """
    def __init__(self, id_dispositivo: str, grabando: bool = False):
        super().__init__(id_dispositivo)
        self.grabando = grabando

    def encender(self):
        self._set_estado(True)
        self.grabando = True
        print(f"ğŸ“· CÃ¡mara {self._get_id()} encendida (grabando).")

    def apagar(self):
        self._set_estado(False)
        self.grabando = False
        print(f"ğŸ“· CÃ¡mara {self._get_id()} apagada.")

    def mostrar_datos(self):
        estado = "Encendida" if self._get_estado() else "Apagada"
        modo = "Grabando" if self.grabando else "En espera"
        print(f"[CamaraSeguridad] ID: {self._get_id()} | Estado: {estado} | Modo: {modo}")
```

```python
# ============================================================
# SensorMovimiento
# ============================================================

class SensorMovimiento(Dispositivo):
    """
    Sensor que detecta movimiento (simulado).
    """
    def __init__(self, id_dispositivo: str, movimiento_detectado: bool = False):
        super().__init__(id_dispositivo)
        self.movimiento_detectado = movimiento_detectado

    def encender(self):
        self._set_estado(True)
        print(f"ğŸ›°ï¸ Sensor {self._get_id()} activado y listo.")

    def apagar(self):
        self._set_estado(False)
        print(f"ğŸ›°ï¸ Sensor {self._get_id()} desactivado.")

    def detectar_movimiento(self, hay_movimiento: bool):
        """Simula la detecciÃ³n de movimiento (solo si estÃ¡ encendido)."""
        if self._get_estado():
            self.movimiento_detectado = hay_movimiento
            mensaje = "Movimiento detectado ğŸš¨" if hay_movimiento else "Sin movimiento"
            print(f"ğŸ›°ï¸ Sensor {self._get_id()} â†’ {mensaje}")
        else:
            print(f"ğŸ›°ï¸ Sensor {self._get_id()} no estÃ¡ encendido. No puede detectar.")

    def mostrar_datos(self):
        estado = "Encendido" if self._get_estado() else "Apagado"
        deteccion = "Detecta movimiento" if self.movimiento_detectado else "Sin movimiento"
        print(f"[SensorMovimiento] ID: {self._get_id()} | Estado: {estado} | {deteccion}")
```

```python
# ============================================================
# CasaInteligente (composiciÃ³n)
# ============================================================

class CasaInteligente:
    """
    Agrupa dispositivos y ejecuta escenas automÃ¡ticas.
    """
    def __init__(self):
        self.dispositivos = []

    def agregar_dispositivo(self, dispositivo: Dispositivo):
        self.dispositivos.append(dispositivo)
        print(f"â• Dispositivo {dispositivo._get_id()} agregado a la casa.")

    def mostrar_todos(self):
        print("\nğŸ“‹ Estado general de los dispositivos:")
        for d in self.dispositivos:
            d.mostrar_datos()

    def ejecutar_escena(self):
        """
        Si algÃºn sensor detecta movimiento â†’ encender luces y cÃ¡maras.
        """
        print("\nğŸš¨ Ejecutando escena automÃ¡tica...")
        hay_movimiento = any(
            isinstance(d, SensorMovimiento) and d.movimiento_detectado
            for d in self.dispositivos
        )
        if hay_movimiento:
            print("ğŸ”” Movimiento detectado: encendiendo luces y cÃ¡maras.")
            for d in self.dispositivos:
                if isinstance(d, (LuzInteligente, CamaraSeguridad)):
                    d.encender()
        else:
            print("âœ… No hay movimiento. No se realizan cambios.")
```

```python
# ============================================================
# PreparaciÃ³n: crear dispositivos y registrarlos en la casa
# ============================================================
# (Ejecuta esta celda para crear y ver que se agregan correctamente)

# Crear instancias
luz_sala = LuzInteligente("L1")                # luz con intensidad por defecto (0)
luz_cuarto = LuzInteligente("L2", intensidad=40)
camara = CamaraSeguridad("C1")
sensor1 = SensorMovimiento("S1")
sensor2 = SensorMovimiento("S2")

# Crear la casa y agregar dispositivos
casa = CasaInteligente()
casa.agregar_dispositivo(luz_sala)
casa.agregar_dispositivo(luz_cuarto)
casa.agregar_dispositivo(camara)
casa.agregar_dispositivo(sensor1)
casa.agregar_dispositivo(sensor2)
```

```python
# ============================================================
# Encender sensores y simular detecciÃ³n de movimiento
# ============================================================
# (Ejecuta esta celda para activar sensores y simular un evento)

sensor1.encender()
sensor2.encender()

# Simulamos que S1 detecta movimiento, S2 no
sensor1.detectar_movimiento(True)
sensor2.detectar_movimiento(False)
```

```python
# ============================================================
# Mostrar estados y ejecutar la escena automÃ¡tica
# ============================================================

# Estado antes de la escena
casa.mostrar_todos()

# Ejecutar escena (si hay movimiento, encenderÃ¡ luces y cÃ¡maras)
casa.ejecutar_escena()

# Estado despuÃ©s de la escena
casa.mostrar_todos()
```

```python
# ============================================================
# Pruebas adicionales 
# ============================================================

# Apagar un sensor y volver a ejecutar escena
sensor1.apagar()
sensor1.detectar_movimiento(True)  # debe indicar que no puede detectar porque estÃ¡ apagado
casa.ejecutar_escena()

# Forzar apagar todas las luces
for d in casa.dispositivos:
    if isinstance(d, LuzInteligente):
        d.apagar()

casa.mostrar_todos()
```
